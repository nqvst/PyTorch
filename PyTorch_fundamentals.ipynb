# Matrix Multiplication

```python
t1 = torch.tensor([
    [1,2,3],
    [4,5,6]
])

t2 = torch.tensor([
    [7,8],
    [9,10],
    [11,12]
])

t1.ndim, t2.ndim
```

```python
t1 * t1 # elementwise multiplication
```

```python
torch.matmul(t1,t2) # == torch.mm == @
```

```python
t3 = torch.arange(100000)
t3.shape
```

# Time comparison
the torch.matmul is orders of magnitude faster at doing matrix multiplicaton than doing it by hand in a for-loop


```python
%%time
value = 0
for i in range(len(t3)):
    value += t3[i] * t3[i]
print(value)
```

```python
%%time
torch.matmul(t3, t3)
```

# Rules for matmul to work
1. the inner dimensions need to match:
- `(2, 3) @ (2, 3)` -> error
- `(2, 3) @ (3, 2)` -> works
    - by inner dimensions we mean the numbers closest to the center of the expression.

2. the resulting tensor will be of the outer dimensions.
- `(2, 3) @ (3, 2)` -> `(2, 2)`
    - the outer dimensions are the starting 2 and the trailing 2 in the expression.

```python
tm1 = torch.rand(2,3) @ torch.rand(3,2)
tm1.shape, tm1
```

```python
tm2 = torch.rand(4,3) @ torch.rand(3,4)
tm2.shape, tm2
```

# Reshaping, stacking, squeezing and unzqueezing tensors
- Reshaping - reshape a tensor
- View - returns a part of a tensor in a specific shape but shares the pointer to the original
- Stacking - combine multiple tensors on top of eachother (vstack, hstack)
- Squeeze - removes all `1` dimensions
- Unsqueeze - add a `1` dimension to a tensor
- Permute - retujrn a view of the input with dimensions swapped in a certain way

```python
x = torch.arange(1., 10.)
x, x.shape
```

## Reshape

```python
# the new shape needs to be reshape(x,y) where x * y = tensor.shape
# int this case shape = 9
# 3 * 3 = 9 ok!
# 2 * 9 = 18 NOT ok!
# x.reshape(2, 3) NOPE
x.reshape(3, 3) # yes!
```

## View

```python
# same rules apply here as with reshape.
# the elements need to match
x.view(3,3)
```

### changing elements in a view also changes the original tensor.
